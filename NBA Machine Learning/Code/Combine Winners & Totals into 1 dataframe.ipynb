```python
import psycopg2
import pandas as pd
import sys
import csv
import sqlalchemy
from sqlalchemy import create_engine
engine = create_engine('sqlite:///:memory:')
```

```python
nba_total_data = pd.read_csv('../data/Totals_data_1990_to_2016.csv')
nba_award_data = pd.read_csv('../data/Transformed_Award_winners_1990_to_2015.csv')
```

```python
Total=nba_total_data.to_sql('Total', engine)
Award = nba_award_data.to_sql('Award', engine)
```

```python
pd.read_sql_query('SELECT * FROM Total', engine)
```

```python
#Left join the two tables together so winners & total stats are on 1 data frame
Merged_df = pd.read_sql_query('SELECT * FROM Total LEFT JOIN Award ON Total.Player = Award.Winner \
AND Total.Season = Award.Year ORDER by Winner', engine)
```

```python
#drop the stuff we dont need
Merged_df = Merged_df.drop(['index','Unnamed: 0', 'Year', 'Winner'], axis=1)
```

```python
#Replace the NaN in Team column with 0s
Merged_df['Team']= Merged_df['Team'].fillna(0)
Merged_df
```

```python
#save to csv
Merged_df.to_csv("../Data/TotalStats&AllTeam_1990_to_2015")
```
